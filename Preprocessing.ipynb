````python
%%bash
for bed in data/*.bed.gz; do
    name="$(basename $bed)"
    name="${name%%.bed.gz}"
    
    echo "$name total"; zcat $bed | grep -v "^#" | sort -k1,1V -k2,2n -k3,3n -u |wc -l;
    echo "$name rev"; zcat $bed | grep -v "^#" | sort -k1,1V -k2,2n -k3,3n -u | grep "reverse_complement=yes" | wc -l;
    echo "$name fwd"; zcat $bed | grep -v "^#" | sort -k1,1V -k2,2n -k3,3n -u | grep "reverse_complement=no" | wc -l;
done | paste - -
````

````python
%%bash
mkdir -p data/stranded/
for bed in data/*.bed.gz; do
    name="$(basename $bed)"
    name="${name%%.bed.gz}"
    
    echo "$name rev"; zcat $bed | grep -v "^#" | sort -k1,1V -k2,2n -k3,3n -u | grep "reverse_complement=yes" | grep -v 'chrUn\|_random\|_alt \|_fix\|chrEBV' | gzip -c > data/stranded/$name.dedup.rev.bed.gz;
    echo "$name fwd"; zcat $bed | grep -v "^#" | sort -k1,1V -k2,2n -k3,3n -u | grep "reverse_complement=no" | grep -v 'chrUn\|_random\|_alt \|_fix\|chrEBV' | gzip -c > data/stranded/$name.dedup.fwd.bed.gz;
    echo "$name rev"; zcat $bed | grep -v "^#" | grep "reverse_complement=yes" | grep -v 'chrUn\|_random\|_alt \|_fix\|chrEBV' | gzip -c > data/stranded/$name.rev.bed.gz;
    echo "$name fwd"; zcat $bed | grep -v "^#" | grep "reverse_complement=no" | grep -v 'chrUn\|_random\|_alt \|_fix\|chrEBV' | gzip -c > data/stranded/$name.fwd.bed.gz;
done
````
